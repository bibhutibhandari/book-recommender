```python
import pandas as pd
```

```python
final_ratings_pt = pd.read_csv("../data/processed/final_ratings_pt.csv")
final_ratings_pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
final_ratings_df =pd.DataFrame(final_ratings_pt).set_index('Book-Title')
```

```python
final_ratings_df
```

```python
similarity_score = cosine_similarity(final_ratings_df)
```

```python
similarity_score[0]
```

```python
np.where(final_ratings_df.index=='1984')[0][0]
```

```python
similarity_df = pd.DataFrame(similarity_score, index=final_ratings_df.index, columns= final_ratings_df.index)
```

```python
similarity_df
```

```python
def book_recommender(book_name):

    similar_books = sorted(list(enumerate(similarity_df[book_name])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_books:
        print(similarity_df.index[i[0]])
```

```python
similarity_df.index
```

```python
book_recommender('The Notebook')
```

```python
book_recommender('Wicked: The Life and Times of the Wicked Witch of the West')
```

```python
book_recommender('Empire Falls')
```
