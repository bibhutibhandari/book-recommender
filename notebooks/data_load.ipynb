```python
import pandas as pd
```

```python
books = pd.read_csv('../data/raw/BX-Books.csv', encoding='latin1', sep=';', on_bad_lines='skip')
books.head()
```

```python
book_ratings = pd.read_csv('../data/raw/BX-Book-Ratings.csv', encoding='latin1',sep=';')
book_ratings.head()
```

```python
users = pd.read_csv('../data/raw/BX-Users.csv', encoding='latin1',sep=';')
users.head()
```

```python
book_ratings.duplicated().sum()
```

```python
books.duplicated().sum()
```

```python
users.duplicated().sum()
```

```python
book_rating_and_name = books.merge(book_ratings, on = 'ISBN')
book_rating_and_name.info()
```

```python
user_count = book_rating_and_name.groupby('User-ID').count()['Book-Rating']
user_count
```

```python
user_count= user_count>250
user_count
```

```python
users_with_ratings = user_count[user_count].index
users_with_ratings
```

```python
temp = book_rating_and_name['User-ID'].isin(users_with_ratings)
ratings_filtered = book_rating_and_name[temp]
ratings_filtered
```

```python
temp_x = ratings_filtered.groupby('Book-Title').count()['Book-Rating']
temp_x = temp_x>50
popularity_ranked_books = temp_x[temp_x].index
popularity_ranked_books
```

```python
popularity_ratings_filtered= ratings_filtered[ratings_filtered['Book-Title'].isin(popularity_ranked_books)]
popularity_ratings_filtered.info()
```

```python
popularity_ratings_filtered['Book-Title'].nunique()
```

```python
popularity_ratings_filtered
```

```python
popularity_ratings_filtered= popularity_ratings_filtered.drop(columns=['Image-URL-S','Image-URL-M','Image-URL-L'])
```
